## Setup

```python
import numpy as np
import iris.utils as utils
```

```python
image_folder = "../images/pasnormalstudios/"
coco_json = "../results/output_coco_dataset.json"
image_id = 12
np.random.seed(3)
```

```python
image_info, reconsturcted_data, image = utils.reconstruct_image_data(coco_json, image_id, image_folder)
```

```python
masks = np.array([item['segmentation'] for item in reconsturcted_data])
mask_quality = np.array([item['predicted_iou'] for item in reconsturcted_data])
point_coords = np.array([item['point_coords'] for item in reconsturcted_data])
bboxes = np.array([np.array(item['bbox']).astype(int) for item in reconsturcted_data])


utils.plot_masks_on_image(
    image, 
    masks, 
    bboxes=bboxes,
    bboxes_on_overlay=False,
    mask_quality=mask_quality, 
    point_coords=point_coords
)
```
